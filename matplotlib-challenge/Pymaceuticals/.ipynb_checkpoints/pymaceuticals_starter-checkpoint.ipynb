```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
md_df = pd.read_csv(mouse_drug_data_to_load)
ct_df = pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
fulldf = pd.merge(md_df, ct_df, how="left", on="Mouse ID")

# Display the data table for preview
fulldf.columns = [c.replace(' ', '_') for c in fulldf.columns]
fulldf.columns = [c.replace('(', '') for c in fulldf.columns]
fulldf.columns = [c.replace(')', '') for c in fulldf.columns]
fulldf
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
meandf = (fulldf.groupby(['Drug','Timepoint']).mean()['Tumor_Volume_mm3'])
# Convert to DataFrame
meandf = pd.DataFrame(meandf)
# Preview DataFrame
meandf.head()
```

```python
semdf = (fulldf.groupby(['Drug','Timepoint']).sem()['Tumor_Volume_mm3'])
semdf = pd.DataFrame(semdf)
semdf.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
meandf = meandf.reset_index()
meanPdf = meandf.pivot(index="Timepoint", columns="Drug")['Tumor_Volume_mm3']
semdf = semdf.reset_index()
semPdf = semdf.pivot(index="Timepoint", columns="Drug")['Tumor_Volume_mm3']
# Preview that Reformatting worked
meanPdf
```

```python
semPdf
```

```python
# Generate the Plot (with Error Bars)
plt.figure(figsize=(8,5))
plt.errorbar(meanPdf.index, meanPdf["Capomulin"], yerr=semPdf["Capomulin"], marker="o", color="r", markersize=5, linestyle="dashed", linewidth=0.5, capsize=3)
plt.errorbar(meanPdf.index, meanPdf["Infubinol"], yerr=semPdf["Infubinol"], marker="x", color="b", markersize=5, linestyle="dashed", linewidth=0.5, capsize=3)
plt.errorbar(meanPdf.index, meanPdf["Ketapril"], yerr=semPdf["Ketapril"], marker="+", color="g", markersize=5, linestyle="dashed", linewidth=0.5, capsize=3)
plt.errorbar(meanPdf.index, meanPdf["Placebo"], yerr=semPdf["Placebo"], marker="^", color="k", markersize=5, linestyle="dashed", linewidth=0.5, capsize=3)


plt.ylim(20,80)
plt.xlim(0,45)
plt.title("Tumor Response to Treatment", fontsize=20)
plt.ylable("Tumor_Volume_mm3", fontsize=14)
plt.xlable("Time (days)", fontsize=14)
plt.grid(linestyle="dashed")
plt.legend(loc="best", fontsize="small", fancybox=True)
plt.savefig("Fig1.png")
```

```python
# Show the Figure
plt.show()
```

![Tumor Response to Treatment](../Images/treatment.png)

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
meanMSdf = (fulldf.groupby(['Drug','Timepoint']).mean()['Metastatic_Sites'])
# Convert to DataFrame
meanMSdf = pd.DataFrame(meanMSdf)
# Preview DataFrame
meanMSdf.head()
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
semMSdf = (fulldf.groupby(['Drug','Timepoint']).sem()['Metastatic_Sites'])
# Convert to DataFrame
semMSdf = pd.DataFrame(semMSdf)
# Preview DataFrame
semMSdf.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
# Minor Data Munging to Re-Format the Data Frames
meanMSdf = meanMSdf.reset_index()
meanMSPdf = meanMSdf.pivot(index="Timepoint", columns="Drug")['Metastatic_Sites']
semMSdf = semMSdf.reset_index()
semMSPdf = semMSdf.pivot(index="Timepoint", columns="Drug")['Metastatic_Sites']
# Preview that Reformatting worked
meanMSPdf
# Preview that Reformatting worked
```

```python
semMSPdf
```

```python
# Generate the Plot (with Error Bars)
plt.figure(figsize=(8,5))
plt.errorbar(meanMSPdf.index, meanMSPdf["Capomulin"], yerr=semMSPdf["Capomulin"], marker="o", color="r", markersize=5, linestyle="dashed", linewidth=0.5, capsize=3)
plt.errorbar(meanMSPdf.index, meanMSPdf["Infubinol"], yerr=semMSPdf["Infubinol"], marker="x", color="b", markersize=5, linestyle="dashed", linewidth=0.5, capsize=3)
plt.errorbar(meanMSPdf.index, meanMSPdf["Ketapril"], yerr=semMSPdf["Ketapril"], marker="+", color="g", markersize=5, linestyle="dashed", linewidth=0.5, capsize=3)
plt.errorbar(meanMSPdf.index, meanMSPdf["Placebo"], yerr=semMSPdf["Placebo"], marker="^", color="k", markersize=5, linestyle="dashed", linewidth=0.5, capsize=3)


plt.ylim(0,4)
plt.xlim(0,45)
plt.title("Metastatic Spread During Treatment", fontsize=20)
plt.ylable("Metastatic Sites", fontsize=14)
plt.xlable("Treatment duration (days)", fontsize=14)
plt.grid(linestyle="dashed")
plt.legend(loc="best", fontsize="small", fancybox=True)
plt.savefig("Fig2.png")
plt.show()
```

![Metastatic Spread During Treatment](../Images/spread.png)

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
scdf = (fulldf.groupby(['Drug','Timepoint']).count()['Tumor_Volume_mm3'])
# Convert to DataFrame
scdf = pd.DataFrame({"Mouse Count": scdf})

# Preview DataFrame
scdf
```

```python
# Minor Data Munging to Re-Format the Data Frames
scdf = scdf.reset_index()
scPdf = scdf.pivot(index="Timepoint", columns="Drug")['Mouse Count']

# Preview that Reformatting worked
scPdf
# Preview the Data Frame
```

```python
# Generate the Plot (Accounting for percentages)
plt.figure(figsize=(8,5))
plt.plot(100 * scPdf["Capomulin"], yerr=scPdf["Capomulin"]/25, "ro", linestyle="dashed", markersize=5, linestyle="dashed", linewidth=0.5)
plt.plot(100 * scPdf["Infubinol"], yerr=scPdf["Infubinol"]/25, "ro",linestyle="dashed", markersize=5, linestyle="dashed", linewidth=0.5)
plt.plot(100 * scPdf["Ketapril"], yerr=scPdf["Ketapril"]/25, "ro",linestyle="dashed", markersize=5, linestyle="dashed", linewidth=0.5)
plt.plot(100 * scPdf["Capomulin"], yerr=scPdf["Placebo"]/25, "ro",linestyle="dashed", markersize=5, linestyle="dashed", linewidth=0.5)

plt.ylim(0,4)
plt.xlim(0,45)
plt.title("Survival During Treatment", fontsize=20)
plt.ylable("Metastatic Sites", fontsize=14)
plt.xlable("Time of day", fontsize=14)
plt.grid(linestyle="dashed")
plt.legend(loc="best", fontsize="small", fancybox=True)
# Save the Figure
plt.savefig("Fig3.png")
plt.show()

# Show the Figure
plt.show()
```

![Metastatic Spread During Treatment](../Images/survival.png)
